```python
import torch
from torch_geometric.data import Data

import graphviz
```

```python
def generate_kmers(string, k):
    """
    Return all kmers of a string
    """
    k_mers = []
    for i in range(len(string) - k + 1):
        k_mers.append(string[i:i + k])

    return k_mers
```

```python
def de_bruijn_graph(kmers):
    nodes = set()
    edge_attr = []
    edge_idx = []
    overlap = len(kmers[0]) - 1

    for i, mer in enumerate(kmers):
        preffix = mer[:overlap]
        suffix = mer[1:]
        nodes.add(preffix)
        nodes.add(suffix)
        edge_attr.append(mer)
        edge_idx.append((preffix, suffix))

    return nodes, edge_attr, edge_idx
```

```python
G = graphviz.Digraph(engine='dot')

sequence = "TAATGCCATGGGATGTT"
kmers = generate_kmers(sequence, 3)
nodes, edge_attr, edge_idx = de_bruijn_graph(kmers)

for node in nodes:
    G.node(node, label=node)

for i, idx in enumerate(edge_idx):
    G.edge(*idx, label=edge_attr[i])

G
```
